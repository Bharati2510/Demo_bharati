# IQR Method

```python
import numpy as np
q1,q3=np.percentile(df["Price"],[25,75])
q1,q3
```

```python
iqr=q3-q1
iqr
```

```python
M=df["Price"].mean()
M
```

```python
UB=M+1.5*iqr
UB
```

```python
LB=M-1.5*iqr
LB
```

```python
df=df[(df["Price"]>LB)&(df["Price"]<UB)]
```

```python
df
```
